```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import pandas as pd
```

```python
df = pd.read_csv('../datasets/wine/wine.data')
```

```python
X = df.iloc[:,1:]
```

```python
y = df.iloc[:, 0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2)
```

```python
minmax_scaler = MinMaxScaler(feature_range=(0,1))
standard_scaler = StandardScaler()

X_train_standard = standard_scaler.fit_transform(X_train)
X_test_standard = standard_scaler.fit_transform(X_test)

X_train_MINMAX = minmax_scaler.fit_transform(X_train)
X_test_MINMAX = minmax_scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score

knn_raw = KNeighborsClassifier(n_neighbors=5)
knn_raw.fit(X_train, y_train)

y_pred_raw = knn_raw.predict(X_test)

acc_raw = accuracy_score(y_test, y_pred_raw)
f1_raw  = f1_score(y_test, y_pred_raw, average="weighted")

print("WITHOUT scaling")
print("Accuracy:", acc_raw)
print("F1-score:", f1_raw)
```

```python
knn_standard = KNeighborsClassifier(n_neighbors=5)
knn_standard.fit(X_train_standard, y_train)

y_pred_standard = knn_standard.predict(X_test_standard)

stand_acc = accuracy_score(y_test, y_pred_standard)
stand_f1  = f1_score(y_test, y_pred_standard, average="weighted")

print("With standard scaling")
print("Accuracy:", stand_acc)
print("F1-score:", stand_f1)
```

```python
knn_minmax = KNeighborsClassifier(n_neighbors=5)
knn_minmax.fit(X_train_MINMAX, y_train)

y_pred_minmax = knn_minmax.predict(X_test_MINMAX)

minmax_acc = accuracy_score(y_test, y_pred_minmax)
minmax_f1 = f1_score(y_test, y_pred_minmax, average="weighted")

print("With minmax scaling")
print("Accuracy:", minmax_acc)
print("F1-score:", minmax_f1)
```
