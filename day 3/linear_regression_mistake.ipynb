```python
# What im basically doing here on forth, its a feature engineering technique i infered when looking at the data, for now 
# the fact that the passenger has specific cabins is not that usefull, knowing that most of them dont even have one,
# so a more intuitive and more meaningfull way to express that data is to check if the passenger either has or doesnt have a cabin
# so i just made a new column with that boolean and deleted the original sparse Cabin column

df['HasCabin'] = df['Cabin'].notna().astype(int)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error



x_train, x_val, y_train, y_val = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.compose import ColumnTransformer

num_features = ['Age', 'Fare', 'SibSp', 'Parch']
cat_features = ['HasCabin','Sex_male', 'Embarked_Q', 'Embarked_S']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features),
        ('cat', 'passthrough', cat_features)
    ]
)
```

```python
x_train_processed = preprocessor.fit_transform(x_train)
x_val_processed   = preprocessor.transform(x_val)


```

```python

model = LinearRegression()
model.fit(x_train_processed, y_train)

print(f"Coeficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")

```

```python
y_pred = model.predict(x_val_processed)
```

```python
mse = mean_squared_error(y_val, y_pred)
```

```python
mse
```

```python
plt.scatter(y_val, y_pred)
plt.plot([y_val.min(), y_val.max()],
         [y_val.min(), y_val.max()])
```
